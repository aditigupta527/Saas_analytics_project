<h3>3. User Acquisition Analysis

```python
####calculating number of customer signups per month
monthly_signups = customers.groupby('signup_month').size()
```

```python
####plotting monthly customer signups trend
monthly_signups.plot(kind='line', title='Monthly Signups')
plt.ylabel('Sign_up Count')
plt.savefig("monthly_signups.png")
plt.show()
```

```python
####filtering records of active users
active_users = activity[activity['is_active'] == True]
```

```python
####calculating monthly active users (MAU)
mau = active_users.groupby('month')['customer_id'].nunique()
```

```python
####plotting monthly active users trend
mau.plot(kind='line', title='Monthly Active Users')
plt.ylabel('No. of active users')
####saving MAU trend visualization
plt.savefig("mau_trend.png")

plt.show()
```

```python
# Churn rate is calculated as the percentage of customers who became inactive in a given month.
# Formula: Churn Rate = Churned Users / Total Users
# Where total users are the sum of active and churned customers for that month.

active_users = activity[activity['is_active'] == True].groupby('month')['customer_id'].nunique()
churned_users = activity[activity['is_active'] == False].groupby('month')['customer_id'].nunique()

churn_rate = churned_users / (active_users + churned_users)
```

```python
# convert churn_rate index to datetime for proper alignment with ARPU
churn_rate.index = pd.to_datetime(churn_rate.index)
```

```python
# plotting The chart shows the monthly churn rate trend. It helps analyze how many customers are leaving the platform over time. A lower churn rate 
# indicates better customer retention and healthier business growth.

churn_rate.plot(kind='line', title='Monthly Churn Rate')
plt.ylabel('Churn Rate')
plt.savefig("Monthly Churn Rate.png")
plt.show()
```

```python
####counting customers acquired by each marketing channel
customers_by_channel = customers.groupby('acquisition_channel')['customer_id'].count()
```

```python
print(customers_by_channel)
```

```python
####calculating total marketing spend by channel
spend_by_channel = marketing.groupby('channel')['spend_amount'].sum()
```

```python
print(spend_by_channel)
```

```python
####calculating customer acquisition cost (CAC) by channel
cac = spend_by_channel / customers_by_channel
```

```python
print(cac)
```

```python
####plotting CAC by marketing channel
cac.plot(kind='bar', title='Customer Acquisition Cost by Channel')

plt.savefig('Customer Acquisition Cost by Channel.png')
plt.show()
```

```python
# ensure month columns are datetime
revenue['month'] = pd.to_datetime(revenue['month'])
activity['month'] = pd.to_datetime(activity['month'])

# calculate MAU again (so it has proper datetime index)
mau = activity[activity['is_active'] == True].groupby('month')['customer_id'].nunique()

# set month as index for revenue
revenue_monthly = revenue.set_index('month')['total_mrr']

# calculating average revenue per user (ARPU)
arpu = revenue_monthly / mau

print(arpu)
```

```python
# plotting arpu
arpu.plot(kind='line', title='Average Revenue Per User (ARPU)')
plt.ylabel('Revenue per User')

plt.savefig('Average Revenue Per User (ARPU)')
plt.show()
```

```python
####calculating customer lifetime value (LTV)
ltv = arpu / churn_rate
```

```python
print(arpu.index)
print(churn_rate.index)
```

```python
print(ltv)
```

```python
####performing cohort retention analysis

# Create cohort month (first activity month for each customer)
activity['cohort_month'] = activity.groupby('customer_id')['month'].transform('min')

# Calculate cohort index (number of months since signup)
activity['cohort_index'] = (
    (activity['month'].dt.year - activity['cohort_month'].dt.year) * 12 +
    (activity['month'].dt.month - activity['cohort_month'].dt.month)
)

# Filter active users and calculate cohort counts
cohort_data = (
    activity[activity['is_active'] == True]
    .groupby(['cohort_month', 'cohort_index'])['customer_id']
    .nunique()
    .reset_index()
)

# Pivot table
cohort_pivot = cohort_data.pivot(
    index='cohort_month',
    columns='cohort_index',
    values='customer_id'
)

# Cohort size
cohort_size = cohort_pivot.iloc[:, 0]

# Retention calculation
retention = cohort_pivot.divide(cohort_size, axis=0).round(3) * 100
```

```python
####visualizing monthly revenue components

revenue.set_index('month')[['new_mrr','expansion_mrr','churned_mrr']].plot(
    kind='bar',
    figsize=(12,6)
)

plt.title('Monthly Revenue Breakdown')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend(['New MRR','Expansion MRR','Churned MRR'])

plt.show()
```

```python
####creating summary table of key SaaS metrics
metrics = pd.DataFrame({
    "month": mau.index,
    "MAU": mau.values,
    "churn_rate": churn_rate.values,
    "ARPU": arpu.values,
    "LTV": ltv.values
})
```

```python
####saving key SaaS metrics for reporting/dashboard
metrics.to_csv("metrics_summary.csv", index=False)

metrics.head()
```
